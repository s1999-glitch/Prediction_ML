```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv('Updated_Alzheimers_Data_Numeric.csv')
```

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.xlabel('Features')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Define features and target
features = data[['Topic_Numeric', 'Question_Numeric']]
target = (data['High_Confidence_Limit'] > 50).astype(int)  # Binary classification based on the threshold
```

```python
# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Train a Random Forest model to get feature importance
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(features_scaled, target)
```

```python
# Extract feature importances
importances = rf_model.feature_importances_
feature_names = ['Topic_Numeric', 'Question_Numeric']
```

```python
# Plot Feature Importance
plt.figure(figsize=(8, 6))
sns.barplot(x=importances, y=feature_names, palette='viridis')
plt.title('Feature Importance Plot')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('feature_importance_plot.png')  # Save the plot as a PNG file
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
# 2. PCA Explained Variance Ratio
# Perform PCA on the scaled features
pca = PCA(n_components=2)  # You can adjust the number of components based on your data
pca.fit(features_scaled)

# Plot the explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.7, color='blue')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('PCA Explained Variance Ratio')
plt.tight_layout()
plt.savefig('pca_explained_variance_ratio.png')
plt.show()
```
